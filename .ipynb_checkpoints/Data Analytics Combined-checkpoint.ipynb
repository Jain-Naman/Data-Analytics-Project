# Loading Matches file

```python
df2=pd.read_csv('C:/Users/DELL/Desktop/PES/DA Project/matches.csv')
```

```python
#data cleaning, error detection in dataset

error=0
for i in range(len(df2['win_by_runs'])):
    if df2['win_by_runs'][i]> 0 and df2['win_by_wickets'][i]>0:
        error+=1
    else:
        pass

if error>0:
    print('Error')
else:
    print('No Error')
```

```python
#cleaning the dataset
df=df2.drop(["id","venue","umpire1","umpire2","umpire3"],axis=1)
```

```python
#Win percentage if team elects to field after winning toss

field_and_match_win=0
for i in range(len(df2['toss_winner'])):
    if (df2['toss_decision'][i] == 'field' and df2['toss_winner'][i]==df2['winner'][i]):
        field_and_match_win +=1

toss_and_match_win_percent=(field_and_match_win/756)*100
print("First field and match win percent is",toss_and_match_win_percent)
```

```python
#Win percentage if team elects to bat after winning toss

bat_and_match_win=0
for i in range(len(df2['toss_winner'])):
    if (df2['toss_decision'][i] == 'bat' and df2['toss_winner'][i]==df2['winner'][i]):
        bat_and_match_win +=1

toss_and_match_win_percent=(bat_and_match_win/756)*100
print("First bat and match win percent is",toss_and_match_win_percent)
```

```python
#Win percentage if team wins toss

toss_and_match_win=0
for i in range(len(df2['toss_winner'])):
    if df2['toss_winner'][i]==df2['winner'][i]:
        toss_and_match_win +=1

toss_and_match_win_percent=(toss_and_match_win/756)*100
print("Percentage of team winning toss wins match", toss_and_match_win_percent,)
```

```python
chennai_match_win=0
for i in range(len(df2['toss_winner'])):
    if (df2['toss_winner'][i]=='Chennai Super Kings' and df2['winner'][i]=='Chennai Super Kings'):
        chennai_match_win +=1

toss_and_match_win_percent=(chennai_match_win/756)*100
print("Toss winner is CSK and match winner is also CSK", toss_and_match_win_percent)
```

```python
#Plot of number of runs in an over

runs_plot=df3.groupby('over')["runs"].sum()

total_runs_plot= sns.barplot(data=df3,x=df3['over'],y=runs_plot)
```

```python
#Plot of number of runs in an over

runs_plot=df3.groupby('over')["runs"].sum()

total_runs_plot= sns.barplot(data=df3,x=df3['over'],y=runs_plot)
```

```python
#Plot of number of runs in an over

wickets_plot=df3.groupby('over')["wickets"].sum()

wicket_plot= sns.lineplot(data=df3,x=df3['over'],y=wickets_plot)
```

```python
#batting first or batting second

batting_order=["Batting First", "Batting Second"]
Compare = [0,0]

for i in range(len(df3['inning'])):
    if(df3['batting_team'][i]==df3['winner'][i]):
        Compare[0]+=1
    else:
        Compare[1]+=1

batting_plot= sns.barplot(data=df3,x=batting_order,y=Compare)
```
